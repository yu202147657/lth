```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# define constants and other global variables
label_size = 28
tick_size = 20
title_size = 40
line_width = 3
```

```python
df = pd.read_csv('mnist_lenet_300_100_IMP.csv', sep='\t')
alpha = df.iloc[:,0]
train_error = df.iloc[:,1]
```

```python
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(alpha, train_error, linewidth=line_width, c='C0') #, marker = 'o'

plt.title("LeNet (MNIST)", fontsize=title_size)
ax.tick_params(axis='both', labelsize=tick_size)

ax.set_xlabel("Interpolation", fontsize=label_size)
ax.set_ylabel("Train Error (%)", fontsize=label_size)

ax.set_ylim([0, 100])

plt.show()
```
